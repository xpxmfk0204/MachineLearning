## hw (1) 

### Least square cost function g(w)를 작성하시오. 교과서 샘플 코드 참조. 

```python
X1_data = x_data.reshape(1,100)
X2_data = x_data.reshape(100,1)
def model(x_data,w):
    a = w[0] + np.dot(X1_data.T, w[1:])
    return a.T

def least_squares(w):
    cost = np.sum((model(x_data,w)-y_data)**2)
    return cost/float(y_data.size)

X1_data
```

## hw (2)

### gradient descent 알고리즘을 이용해서 최적의 weight vector w를 구하시오. 

```python
from autograd import grad 
import autograd.numpy as np 
import matplotlib.pyplot as plt

# gradient descent function 
def gradient_descent(g,alpha,max_its,w):
    # compute gradient module using autograd
    gradient = grad(g)

    # run the gradient descent loop
    weight_history = [w] # weight history container
    cost_history = [g(w)] # cost function history container
    for k in range(max_its):
        # evaluate the gradient
        grad_eval = gradient(w)

        # take gradient descent step
        w = w - alpha*grad_eval
        
        # record weight and cost
        weight_history.append(w)
        cost_history.append(g(w))
    return weight_history,cost_history

# run gradient descent to minimize g(w)
g = least_squares; 
w = 0.1*np.random.randn(2,1); 
max_its = y_data.size; 
alpha_choice = 10**(-2);
w_history, cost_history = gradient_descent(g,alpha_choice,max_its,w)

plt.plot(cost_history)
plt.title('Cost function history plot')
plt.xlabel('Step (k)')
plt.ylabel('g(w)')
plt.show()

w1 = []
w2 = []
[w1.append(a[0]) for a in w_history]
[w2.append(a[1]) for a in w_history]

plt.plot(w1, w2, 'o')
plt.title('w1 w2 values')
plt.xlabel('$w_1$')
plt.ylabel('$w_2$')
plt.show()
print('w개수: ', len(w_history))
```
